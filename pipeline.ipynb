# Gnration de lgende

```python
def load_image(image_path):
    """
    La fonction load_image a pour entre le chemin d'une image et pour sortie un couple
    contenant l'image traite ainsi que son chemin d'accs.
    La fonction load_image effectue les traitement suivant:
        1. Chargement du fichier correspondant au chemin d'accs image_path
        2. Dcodage de l'image en RGB.
        3. Redimensionnement de l'image en taille (299, 299).
        4. Normalisation des pxels de l'image entre -1 et 1
    """
    # Lecture du fichier image
    img = tf.io.read_file(image_path)
    # Decodage de l'image en RGB
    img = tf.image.decode_jpeg(img, channels=3)
    # Redimensionnement de l'image en taille (299, 299)
    img = tf.image.resize(img, (299, 299))
    # Normalisation des pxels de l'image entre -1 et 1
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    return img, image_path
```

```python
embedding_dim = 256
units = 512 # Taille de la couche cache dans le RNN
top_k = 5000 # Nombre de mots  utiliser dans le vocabulaire
vocab_size = top_k + 1
max_length = 35 # Nombre maximum de mots dans une lgende
attention_features_shape = 64 # Taille de la couche d'attention
```

```python
class CNN_Encoder(tf.keras.Model):
    # Comme les images sont dj prtraits par InceptionV3 est reprsent sous forme compacte
    # L'encodeur CNN ne fera que transmettre ces caractristiques  une couche dense
    def __init__(self, embedding_dim):
        super(CNN_Encoder, self).__init__()
        # forme aprs fc == (batch_size, 64, embedding_dim)
        self.fc = tf.keras.layers.Dense(embedding_dim)

    def call(self, x):
        x = self.fc(x)
        x = tf.nn.relu(x)
        return x
```

```python
class BahdanauAttention(tf.keras.Model):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = tf.keras.layers.Dense(units)
        self.W2 = tf.keras.layers.Dense(units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, features, hidden):
        # features(CNN_encoder output) forme == (batch_size, 64, embedding_dim)

        # forme de la couche cache == (batch_size, hidden_size)
        hidden_with_time_axis = tf.expand_dims(hidden, 1)

        attention_hidden_layer = (tf.nn.tanh(self.W1(features) +
                                             self.W2(hidden_with_time_axis)))

        # Cela vous donne un score non normalis pour chaque caractristique de l'image.
        score = self.V(attention_hidden_layer)

        attention_weights = tf.nn.softmax(score, axis=1)

        context_vector = attention_weights * features
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector, attention_weights


class RNN_Decoder(tf.keras.Model):
    def __init__(self, embedding_dim, units, vocab_size):
        super(RNN_Decoder, self).__init__()
        self.units = units

        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.units,
                                       return_sequences=True,
                                       return_state=True,
                                       recurrent_initializer='glorot_uniform')
        #Couche dense qui aura pour entre la sortie du GRU
        self.fc1 = tf.keras.layers.Dense(self.units)
        # Dernire couche dense
        self.fc2 = tf.keras.layers.Dense(vocab_size)

        self.attention = BahdanauAttention(self.units)

    def call(self, x, features, hidden):
        # L'attention est defini par un modle a part
        context_vector, attention_weights = self.attention(features, hidden)
        # Passage du mot courant  la couche embedding
        x = self.embedding(x)
        # Concatnation
        x = tf.concat([tf.expand_dims(context_vector, 1), x], axis=-1)

        # Passage du vecteur concatn  la gru
        output, state = self.gru(x)

        # Couche dense
        y = self.fc1(output)

        y = tf.reshape(y, (-1, x.shape[2]))

        # Couche dense
        y = self.fc2(y)

        return y, state, attention_weights

    def reset_state(self, batch_size):
        return tf.zeros((batch_size, self.units))
```

```python
# Cration de l'encodeur
encoder = CNN_Encoder(embedding_dim)
# Cration du dcodeur
decoder = RNN_Decoder(embedding_dim, units, vocab_size)
```

```python
def load_image(image_path):
    """
    La fonction load_image a pour entre le chemin d'une image et pour sortie un couple
    contenant l'image traite ainsi que son chemin d'accs.
    La fonction load_image effectue les traitement suivant:
        1. Chargement du fichier correspondant au chemin d'accs image_path
        2. Dcodage de l'image en RGB.
        3. Redimensionnement de l'image en taille (299, 299).
        4. Normalisation des pxels de l'image entre -1 et 1
    """
    # Lecture du fichier image
    img = tf.io.read_file(image_path)
    # Decodage de l'image en RGB
    img = tf.image.decode_jpeg(img, channels=3)
    #img = image_path
    # Redimensionnement de l'image en taille (299, 299)
    img = tf.image.resize(img, (299, 299))
    # Normalisation des pxels de l'image entre -1 et 1
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    return img, image_path
```

```python
image_model = tf.keras.applications.InceptionV3(include_top=False,
                                                weights='imagenet')
# Creation d'une variable qui sera l'entre du nouveau modle de pre-traitement d'images
new_input = image_model.input
# rcuprer la dernire couche cach qui contient l'image en representation compacte
hidden_layer = image_model.layers[-1].output

# Modle qui calcule une representation dense des images avec InceptionV3
image_features_extract_model = tf.keras.Model(new_input, hidden_layer)
```

```python
# Load the JSON string from the file
with open('Captionning/tokenizer.json', 'r', encoding='utf-8') as f:
    tokenizer_json = json.load(f)

# Recreate the tokenizer from the JSON string
tokenizer = tokenizer_from_json(tokenizer_json)
```

```python
def evaluate(image):
    attention_plot = np.zeros((max_length, attention_features_shape))

    hidden = decoder.reset_state(batch_size=1)

    temp_input = tf.expand_dims(load_image(image)[0], 0)
    img_tensor_val = image_features_extract_model(temp_input)
    img_tensor_val = tf.reshape(img_tensor_val, (img_tensor_val.shape[0], -1, img_tensor_val.shape[3]))

    features = encoder(img_tensor_val)

    dec_input = tf.expand_dims([tokenizer.word_index['<start>']], 0)
    result = []

    for i in range(max_length):
        predictions, hidden, attention_weights = decoder(dec_input, features, hidden)

        attention_plot[i] = tf.reshape(attention_weights, (-1, )).numpy()

        predicted_id = tf.random.categorical(predictions, 1)[0][0].numpy()
        result.append(tokenizer.index_word[predicted_id])

        if tokenizer.index_word[predicted_id] == '<end>':
            return result, attention_plot

        dec_input = tf.expand_dims([predicted_id], 0)

    attention_plot = attention_plot[:len(result), :]
    return result, attention_plot
```

```python
# Fonction permettant la reprsentation de l'attention au niveau de l'image
def plot_attention(image, result, attention_plot):
    temp_image = np.array(Image.open(image))

    fig = plt.figure(figsize=(10, 10))

    len_result = len(result)
    for l in range(len_result):
        temp_att = np.resize(attention_plot[l], (8, 8))
        ax = fig.add_subplot(len_result//2, len_result//2, l+1)
        ax.set_title(result[l])
        img = ax.imshow(temp_image)
        ax.imshow(temp_att, cmap='gray', alpha=0.6, extent=img.get_extent())

    plt.tight_layout()
    plt.show()
```

```python
# Prendre en entre une seule photo avec le chemin donn manuellement et obtenir la description
# Charger l'image
tf_img = tf.keras.preprocessing.image.load_img(image, target_size=(299, 299))

# image_debruitee_redimensionnee = tf.image.resize(image_debruitee, (299, 299))
# print(f"Forme de l'image redimensionne : {image_debruitee_redimensionnee.shape}")
# tf_img = tf.keras.preprocessing.image.img_to_array(image_debruitee)
# Obtenir la lgende
result, attention_plot = evaluate(image)

# Supprimer <start> et <end> de la lgende
result = result[:-1]

# Afficher l'image et la lgende
temp_image = np.array(Image.open(image))
plt.imshow(temp_image)
plt.title(' '.join(result))
plt.axis('off')
plt.show()

plot_attention(image, result, attention_plot)
```
